# Préparation de la donnée

```python
# On selectionne uniquement les pays

def country_selection(data):
    '''
    Selectionne uniquement les pays dans le Dataframe des données de la banque mondiale
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    Return:
    ------
    data_country : pd.DataFrame

    '''
    country = data_education['Country Name'].unique()[25:]
    data_country = data.loc[data_education['Country Name'].isin(country)]
    return data_country
```

```python
# On selectionne uniquement les indicateurs qui nous interresse

def indicator_selection(data):
    '''
    Selectionne uniquement les indicateurs qui nous interresse dans le Dataframe des données de la banque mondiale
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    Return:
    ------
    data_indicator : pd.DataFrame

    '''
    indicator = ['SP.POP.TOTL','SP.POP.GROW','SP.POP.1524.TO.UN',
                 'SP.SEC.TOTL.IN','SP.SEC.UTOT.IN','SP.TER.TOTL.IN',
                 'IT.NET.USER.P2','IT.CMP.PCMP.P2','NY.GDP.PCAP.PP.CD']
    data_indicator = data.loc[data_education['Indicator Code'].isin(indicator)]
    return data_indicator
```

```python
# Avant de selectionner les années, on regarde à nouveau la donnée manquante maintenant que l'on a filtrer
# les indicateurs

data_verif = indicator_selection(country_selection(data_education))
print(round(data_verif.isna().sum().sort_values(ascending = False)/len(data_verif),3).to_string())
```

```python
# Selectionner les années entre 2005 et 2015 semble un bon choix ( pas trop de données manquante )
# On selectionne uniquement les colonnes dont on a besoin (On a choisis de garder uniquement les années entre 2005 et 2015)

column = ['Country Name', 'Country Code', 'Indicator Name',
          'Indicator Code',
          '2005','2006', '2007', '2008', '2009', '2010', 
          '2011', '2012', '2013', '2014','2015']

def column_selection(data,column):
    '''
    Selectionne uniquement les colonnes qui nous interresse dans le Dataframe des données de la banque mondiale
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    column : list
    
    Return:
    ------
    data_selected : pd.DataFrame

    '''    
    data_selected = data[column]
    return data_selected
```

```python
# On crée une colonne pour la moyenne
# Non utilisé

year = ['2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013','2014', '2015']

def data_mean(data,year):
    '''
    Crée une colonne correspondant à la moyenne de chaque indicateurs pour chaque pays
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    year : list
    
    Return:
    ------
    data : pd.DataFrame

    '''        

    data['mean'] = data[year].mean(axis=1)
    return data
```

```python
# On crée une colonne pour récupérer la dernière valeur qui n'est pas un NaN pour chaque ligne

def last_line(row):
    '''
    Récupère le dernière élement qui n'est pas un NaN dans une pandas series
    
    Parameters :
    ----------
    row : pd.Series
    
    Return:
    ------
    0 : int

    row[row.last_valid_index()] : float

    '''   
    if row.last_valid_index() is None:
        return 0
    else:
        return row[row.last_valid_index()]
    
def data_last_line(data,year):
    '''
    Récupère le dernière indicateurs renseigné (qui n'est pas un NaN) pour chaque ligne
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    year : list
    
    Return:
    ------
    data : pd.DataFrame

    '''    

    data['last_line'] = data[year].apply(last_line,axis=1)
    return data
```

```python
# On fait une fonction pour executer les 3 selections en même temps et la création du mean

def data_selection_last_line(data,year,column):
    '''
    Crée une colonne correspondant à la moyenne de chaque indicateurs pour chaque pays
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    year : list
    
    column : list
    
    Return:
    ------
    data_last_line : pd.DataFrame

    ''' 
    return data_last_line(indicator_selection(country_selection(column_selection(data,column))),year)
```

```python
# Le premier filtre sera realisé sur les pays avec une population totale >= 5 000 000

def filter_population(data):
    '''
    On garde les pays avec une population totale >= 5 000 000
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    Return:
    ------
    data_population : pd.DataFrame

    ''' 
    country_selected = data[(data['Indicator Code']=='SP.POP.TOTL') 
                                 & (data['last_line'] >= 5_000_000)]['Country Name'].unique()
    data_population = data[data['Country Name'].isin(country_selected)]
    return data_population
```

```python
# Le deuxième filtre est sur le PIB par habitant. Il doit être au moins 80% de celui de la France

def filter_GDP(data):
    '''
    On garde les pays avec un PIB >= 80% celui de la France
    
    Parameters :
    ----------
    data : pd.DataFrame
    
    Return:
    ------
    data_gdp : pd.DataFrame

    ''' 
    french_gdp = data[(data['Country Name']=='France') 
                          & (data['Indicator Code']== 'NY.GDP.PCAP.PP.CD')]['last_line'].iloc[0]
    gdp_selected = data[(data['Indicator Code']=='NY.GDP.PCAP.PP.CD') 
                                     & (data['last_line'] >= french_gdp*0.8)]['Country Name'].unique()
    data_gdp = data[data['Country Name'].isin(gdp_selected)]
    
    return data_gdp
    
```

```python
# On va créer la moyenne des classements
# On commence par extraire le numéros du classement pour chaque pays pour chaque indicateur

indice = ['index','Country Name']

secondaire = data_ready[data_ready['Indicator Code']=='SP.SEC.TOTL.IN'] \
    .sort_values(by='last_line',ascending=False).reset_index(drop=True).reset_index()[indice]
secondaire['classement_secondaire'] = secondaire['index'].apply(lambda x : x+1)
secondaire['index'] = secondaire['index'].apply(lambda x : x+1)

tertiaire = data_ready[data_ready['Indicator Code']=='SP.TER.TOTL.IN'] \
    .sort_values(by='last_line',ascending=False).reset_index(drop=True).reset_index()[indice]
tertiaire['classement_tertiaire'] = tertiaire['index'].apply(lambda x : x+1)
tertiaire['index'] = tertiaire['index'].apply(lambda x : x+1)

internet = data_ready[data_ready['Indicator Code']=='IT.NET.USER.P2'] \
    .sort_values(by='last_line',ascending=False).reset_index(drop=True).reset_index()[indice]
internet['classement_internet'] = internet['index'].apply(lambda x : x+1)
internet['index'] = internet['index'].apply(lambda x : x+1)

portable = data_ready[data_ready['Indicator Code']=='IT.CMP.PCMP.P2'] \
    .sort_values(by='last_line',ascending=False).reset_index(drop=True).reset_index()[indice]
portable['classement_ordinateur_portable'] = portable['index'].apply(lambda x : x+1)
portable['index'] = portable['index'].apply(lambda x : x+1)

population = data_ready[data_ready['Indicator Code']=='SP.POP.TOTL'] \
    .sort_values(by='last_line',ascending=False).reset_index(drop=True).reset_index()[indice]
population['classement_population'] = population['index'].apply(lambda x : x+1)
population['index'] = population['index'].apply(lambda x : x+1)

PIB = data_ready[data_ready['Indicator Code']=='NY.GDP.PCAP.PP.CD'] \
    .sort_values(by='last_line',ascending=False).reset_index(drop=True).reset_index()[indice]
PIB['classement_internet'] = PIB['index'].apply(lambda x : x+1)
PIB['index'] = PIB['index'].apply(lambda x : x+1)
```

```python
# On crée un dataframe avec le classement des pays

data_classement = pd.concat([secondaire, tertiaire, internet,portable,population,PIB], ignore_index=True, sort=False).groupby('Country Name').mean()\
    .sort_values(by='index').rename(columns={'index':'classement_moyen'})

data_classement
```

```python
# On va créer le classement moyen
# Au final, j'ai préféré le dataframe au graphique

classement = pd.concat([secondaire, tertiaire, internet,portable,population,PIB], ignore_index=True, sort=False)\
                .groupby('Country Name').mean().reset_index()

y_label = classement.sort_values(by='index')['Country Name']
y_pos = np.arange(len(y_label))

performance = classement.sort_values(by='index')['index']
plt.title('classement moyen des indicateurs')
plt.barh(y_label,performance)
plt.savefig('classement moyen des rangs',bbox_inches='tight');
```
